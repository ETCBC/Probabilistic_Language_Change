```python
import collections
from itertools import chain
import pandas as pd

countData = dict()
for book, lists in testdata.items():
    countData[book] = collections.Counter(list(chain(*lists))) 
df =pd.DataFrame(countData).fillna(0)
```

```python
wil = collections.defaultdict(dict)
mannWhit = collections.defaultdict(dict)
for book1 in df.keys():
    for book2 in df.keys():   
        wil[book1][book2]     = wilcoxon(df[book1],df[book2], zero_method="zsplit", correction=False)[1] # Paired
        # These tests are for median.
        mannWhit[book1][book2]=mannwhitneyu(df[book1],df[book2], False)[1]# Tests median
```
